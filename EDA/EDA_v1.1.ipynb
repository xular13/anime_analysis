# **Missing Values in `anime_details.csv`**
| Column              | Missing Count | Handling Strategy |
|---------------------|--------------|-------------------|
| `main_picture`      | 10           | Keep as NaN (not critical) |
| `start_date`        | 160          | Fill using `start_season` or mark as "Unknown" |
| `end_date`          | 410          | If `status` is "airing", mark as "Ongoing" |
| `synopsis`          | 269          | Replace with `"No synopsis available"` |
| `mean`             | 698          | Fill using median score of similar `media_type` |
| `rank`             | 1693         | Do **not** fill, can be recomputed later |
| `genres`           | 21           | Replace with `"Unknown"` |
| `start_season`     | 727          | Infer from `start_date` if possible |
| `source`          | 674          | Replace with `"Unknown"` |

---



```python

# Convert 'start_date' and 'end_date' to datetime format safely
anime_details['start_date'] = pd.to_datetime(anime_details['start_date'], errors='coerce', format='mixed')
anime_details['end_date'] = pd.to_datetime(anime_details['end_date'], errors='coerce', format='mixed')

# Fixing 'mean' missing values (handling cases where all values are NaN in a group)
anime_details['mean'] = anime_details.groupby('media_type')['mean'].transform(
    lambda x: x.fillna(x.median()) if not x.isna().all() else x.fillna(anime_details['mean'].median())
)

# Extract year from 'start_season' safely
def extract_year(season):
    """Extract year from start_season JSON-like data."""
    try:
        season_dict = ast.literal_eval(season)
        return season_dict.get('year', None)
    except (ValueError, SyntaxError):
        return None

anime_details['start_year'] = anime_details['start_season'].apply(lambda x: extract_year(x) if pd.notna(x) else None)

# Fill missing start_date with January 1st of the extracted year
anime_details.loc[anime_details['start_date'].isna() & anime_details['start_year'].notna(), 'start_date'] = anime_details['start_year'].apply(lambda x: pd.Timestamp(f"{int(x)}-01-01") if pd.notna(x) else None)

# Fix 'end_date' issue - set ongoing anime to NaT instead of string
anime_details.loc[(anime_details['status'] == 'airing') & (anime_details['end_date'].isna()), 'end_date'] = pd.NaT

# Fill missing 'source', 'genres', and 'synopsis' safely using .loc[]
anime_details.loc[anime_details['source'].isna(), 'source'] = "Unknown"
anime_details.loc[anime_details['genres'].isna(), 'genres'] = "Unknown"
anime_details.loc[anime_details['synopsis'].isna(), 'synopsis'] = "No synopsis available"
anime_details['start_date'].fillna(pd.NaT, inplace=True)
anime_details['end_date'].fillna(pd.NaT, inplace=True)
anime_details['start_season'].fillna(pd.NaT, inplace=True)

# Handling 'rank': Since it depends on 'mean', we recalculate it
anime_details['rank'] = anime_details['mean'].rank(method="min", ascending=False)

# Convert 'rank' to integer, keeping NaN values as they are
anime_details['rank'] = anime_details['rank'].astype('Int64')

# Check missing values after final fixes
missing_values_final = anime_details.isnull().sum()

missing_values_final

# Drop 'start_year' as it was only used for filling missing values
anime_details.drop(columns=['start_year'], inplace=True)

# Check missing values after fixes
missing_values_fixed = anime_details.isnull().sum()

missing_values_fixed
```

# **Step 3: Parsing JSON-Like Fields**
We successfully extracted structured values from JSON-like string columns:

| Column              | **Before** (JSON-like format) | **After Parsing** (Structured List) |
|---------------------|--------------------------------|------------------------------------|
| `genres`           | `'[{"id": 4, "name": "Comedy"}, {"id": 35, "name": "Harem"}]'` | `['Comedy', 'Harem']` |
| `studios`          | `'[{"id": 73, "name": "TMS Entertainment"}]'` | `['TMS Entertainment']` |
| `alternative_titles` | `'{"synonyms": ["Himouto! Umaru-chan 2nd Season"], "en": "My Two-Faced Little Sister"}'` | `['Himouto! Umaru-chan 2nd Season', 'My Two-Faced Little Sister']` |
| `related_anime`    | `'[{"node": {"id": 40807, "title": "Ore wo Suki nano wa Omae dake ka yo: Oretachi no Game Set"}}]'` | `40807` |

## **Key Fixes**
✅ Extracted `genres` and `studios` as **lists of names**  
✅ Cleaned `alternative_titles` for **easier use**  
✅ Extracted related anime titles  




```python
import ast

# Function to safely parse JSON-like strings
def safe_parse_json(value):
    """Parses a JSON-like string safely and returns a list of values."""
    try:
        parsed_value = ast.literal_eval(value)  # Convert string to Python object
        if isinstance(parsed_value, list):  # Ensure it's a list
            return [entry['name'] for entry in parsed_value if isinstance(entry, dict) and 'name' in entry]
        elif isinstance(parsed_value, dict):  # Handle cases where it's a dictionary
            return list(parsed_value.values())
        return parsed_value  # Return as is if neither list nor dict
    except (ValueError, SyntaxError):
        return value  # Return original if parsing fails

def extract_related_anime_ids(value):
    """Parses a JSON-like related_anime field and extracts only the anime IDs."""
    try:
        parsed_value = ast.literal_eval(value)  # Convert string to Python object
        if isinstance(parsed_value, list):  # Ensure it's a list
            return [entry['node']['id'] for entry in parsed_value if isinstance(entry, dict) and 'node' in entry and 'id' in entry['node']]
        return []  # Return empty list if not in expected format
    except (ValueError, SyntaxError):
        return []  # Return empty list if parsing fail

# Apply parsing to JSON-like columns
anime_details['genres'] = anime_details['genres'].apply(safe_parse_json)
anime_details['studios'] = anime_details['studios'].apply(safe_parse_json)
anime_details['alternative_titles'] = anime_details['alternative_titles'].apply(safe_parse_json)
anime_details['related_anime'] = anime_details['related_anime'].apply(extract_related_anime_ids)


# Check transformed data
anime_details[['genres', 'studios', 'alternative_titles', 'related_anime']].head()
```

# **Step 7: Comparing Short vs. Long Anime Ratings 📏**

```python
# Define categories based on episode count
def categorize_anime_length(episodes):
    if pd.isna(episodes) or episodes == 0:
        return "Unknown"
    elif episodes == 1:
        return "Single Episode/Film"
    elif episodes <= 12:
        return "Short (1-12 Episodes)"
    elif episodes <= 24:
        return "Medium (13-24 Episodes)"
    elif episodes <= 50:
        return "Long (25-50 Episodes)"
    else:
        return "Very Long (50+ Episodes)"

# Apply categorization
anime_details['episode_category'] = anime_details['num_episodes'].apply(categorize_anime_length)

# Group by episode category and compute mean ratings
length_ratings = anime_details.groupby('episode_category')['mean'].agg(['mean', 'count']).dropna()

# Sorting for better visualization
length_ratings = length_ratings.sort_values(by='mean', ascending=False)

# Visualization: Ratings by Anime Length
plt.figure(figsize=(10, 5))
sns.barplot(y=length_ratings.index, x=length_ratings['mean'], palette="coolwarm")

# Improve readability
plt.xlabel("Average Mean Score", fontsize=12)
plt.ylabel("Episode Length Category", fontsize=12)
plt.title("Average Anime Ratings by Episode Length", fontsize=14, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()

# Display numerical summary
length_ratings
```

# **Step 13: User-Genre Preferences Analysis 🎭**

```python

from collections import defaultdict



# Create a genre dictionary for fast lookup
anime_genre_dict = anime_details.set_index('id')['genres'].to_dict()

# Dictionary to store unique genre counts per user
user_genre_count_dict = defaultdict(set)

# Process only the small batch
for _, row in animelist.iterrows():
    anime_id = row['anime_id']
    user_id = row['user_id']

    # Get genres for this anime
    genres = anime_genre_dict.get(anime_id, [])
    if isinstance(genres, list):
        user_genre_count_dict[user_id].update(genres)

# Convert to DataFrame
user_genre_counts_final = pd.DataFrame({
    'user_id': list(user_genre_count_dict.keys()),
    'unique_genres': [len(genres) for genres in user_genre_count_dict.values()]
})

# Print sample results
print(user_genre_counts_final.head())
```

```python
mean_value = user_genre_counts_final['unique_genres'].mean()
median_value = user_genre_counts_final['unique_genres'].median()
percentiles = user_genre_counts_final['unique_genres'].quantile([0.25, 0.75])

# Enhanced Visualization
plt.figure(figsize=(10, 5))
sns.histplot(user_genre_counts_final['unique_genres'], bins=20, kde=True, color='purple', edgecolor='black', alpha=0.6)

# Add vertical lines for mean, median, and quartiles
plt.axvline(mean_value, color='blue', linestyle='dashed', linewidth=2, label=f"Mean: {mean_value:.1f}")
plt.axvline(median_value, color='green', linestyle='dashed', linewidth=2, label=f"Median: {median_value:.1f}")
plt.axvline(percentiles[0.25], color='red', linestyle='dotted', linewidth=1.5, label=f"25th Percentile: {percentiles[0.25]:.1f}")
plt.axvline(percentiles[0.75], color='red', linestyle='dotted', linewidth=1.5, label=f"75th Percentile: {percentiles[0.75]:.1f}")

# Improve readability
plt.xlabel("Number of Unique Genres Watched", fontsize=12)
plt.ylabel("Number of Users", fontsize=12)
plt.title("Distribution of Unique Genres Watched per User", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(alpha=0.3)

# Show plot
plt.show()
```
